I want to keep the upper level groupings of government payouts (2000) level.

I do not want the full breakdown. (2X00) is the mold I am looking for.

```python
# Logic to break down code into the correct lines
index_filt = (transfer['LineCode']>2000) & (transfer['LineCode']<3000) & (transfer['LineCode']%100==0)
```

```python
transfer = transfer[index_filt]
```
