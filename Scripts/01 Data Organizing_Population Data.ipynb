# Shape DataFrame and Delete Columns

```python
# Remove Rural-urban_Continuum columns
pop = pop[pop.columns.drop(list(pop.filter(regex='Rural-urban_Continuum')))]
```

```python
pop.columns.tolist()
```

```python
# Remove GQ_ESTIMATES
pop = pop[pop.columns.drop(list(pop.filter(regex='GQ_ESTIMATES')))]
```

```python
# Remove R_ (rates)
pop = pop[pop.columns.drop(list(pop.filter(regex='R_')))]
```

```python
# Remove RESIDUAL
pop = pop[pop.columns.drop(list(pop.filter(regex='RESIDUAL')))]
```

```python
# Remove NET_MIG
pop = pop[pop.columns.drop(list(pop.filter(regex='NET_MIG')))]
```

```python
# Remove remaining unnecessary columns
pop = pop[pop.columns.drop(['Urban_Influence_Code_2003',
 'Urban_Influence_Code_2013',
 'Economic_typology_2015',
 'CENSUS_2010_POP',
 'ESTIMATES_BASE_2010'])]
```

```python
pop.shape
```

I need to break the DataFrame into multiple DataFrames in order to use the `.melt` function and then bring them back together into a shape that I want.

```python
#Population Estimate
pop_est_list = ['FIPStxt','State','Area_Name','POP_ESTIMATE_2010',
 'POP_ESTIMATE_2011','POP_ESTIMATE_2012','POP_ESTIMATE_2013',
 'POP_ESTIMATE_2014','POP_ESTIMATE_2015','POP_ESTIMATE_2016',
 'POP_ESTIMATE_2017','POP_ESTIMATE_2018','POP_ESTIMATE_2019']

pop_est = pop.filter(pop_est_list)
```

```python
pop_est
```

```python
# remove POP_ESTIMATE_ from columns
pop_est.columns = [col.replace('POP_ESTIMATE_','') for col in pop_est.columns]
```

```python
pop_est = pd.melt(pop_est,id_vars=['FIPStxt','State','Area_Name'],
                 var_name = 'Year',value_name = 'Pop_Est')
```

```python
#N_POP
n_pop_list = ['FIPStxt','State','Area_Name','N_POP_CHG_2010',
 'N_POP_CHG_2011','N_POP_CHG_2012','N_POP_CHG_2013',
 'N_POP_CHG_2014','N_POP_CHG_2015','N_POP_CHG_2016',
 'N_POP_CHG_2017','N_POP_CHG_2018','N_POP_CHG_2019']
n_pop = pop.filter(n_pop_list)
```

```python
# remove N_POP_CHG_ from columns
n_pop.columns = [col.replace('N_POP_CHG_','') for col in n_pop.columns]
```

```python
n_pop = pd.melt(n_pop,id_vars=['FIPStxt','State','Area_Name'],
                 var_name = 'Year',value_name = 'Pop_Change')
```

```python
#Births
birth_list = ['FIPStxt','State','Area_Name','Births_2010',
 'Births_2011','Births_2012','Births_2013',
 'Births_2014','Births_2015','Births_2016',
 'Births_2017','Births_2018','Births_2019']
birth = pop.filter(birth_list)
```

```python
# remove Births_ from columns
birth.columns = [col.replace('Births_','') for col in birth.columns]
```

```python
birth = pd.melt(birth,id_vars=['FIPStxt','State','Area_Name'],
                 var_name = 'Year',value_name = 'Births')
```

```python
#Deaths
death_list = ['FIPStxt','State','Area_Name','Deaths_2010',
 'Deaths_2011','Deaths_2012','Deaths_2013',
 'Deaths_2014','Deaths_2015','Deaths_2016',
 'Deaths_2017','Deaths_2018','Deaths_2019']
death = pop.filter(death_list)

# remove Deaths_ from columns
death.columns = [col.replace('Deaths_','') for col in death.columns]

# reduce into rows
death = pd.melt(death,id_vars=['FIPStxt','State','Area_Name'],
                 var_name = 'Year',value_name = 'Deaths')
```

```python
#Natural Increases
nat_inc_list = ['FIPStxt','State','Area_Name','NATURAL_INC_2010',
 'NATURAL_INC_2011','NATURAL_INC_2012','NATURAL_INC_2013',
 'NATURAL_INC_2014','NATURAL_INC_2015','NATURAL_INC_2016',
 'NATURAL_INC_2017','NATURAL_INC_2018','NATURAL_INC_2019']
nat_inc = pop.filter(nat_inc_list)

# remove NATURAL_INC from columns
nat_inc.columns = [col.replace('NATURAL_INC_','') for col in nat_inc.columns]

# reduce into rows
nat_inc = pd.melt(nat_inc,id_vars=['FIPStxt','State','Area_Name'],
                 var_name = 'Year',value_name = 'Nat_Inc')
```

```python
#International Migrations
int_mig_list = ['FIPStxt','State','Area_Name','INTERNATIONAL_MIG_2010',
 'INTERNATIONAL_MIG_2011','INTERNATIONAL_MIG_2012','INTERNATIONAL_MIG_2013',
 'INTERNATIONAL_MIG_2014','INTERNATIONAL_MIG_2015','INTERNATIONAL_MIG_2016',
 'INTERNATIONAL_MIG_2017','INTERNATIONAL_MIG_2018','INTERNATIONAL_MIG_2019']
int_mig = pop.filter(int_mig_list)

# remove INTERNATIONAL_MIG from columns
int_mig.columns = [col.replace('INTERNATIONAL_MIG_','') for col in int_mig.columns]

# reduce into rows
int_mig = pd.melt(int_mig,id_vars=['FIPStxt','State','Area_Name'],
                 var_name = 'Year',value_name = 'Int_Mig')
```

```python
#Domestic Migrations
dom_mig_list = ['FIPStxt','State','Area_Name','DOMESTIC_MIG_2010',
 'DOMESTIC_MIG_2011','DOMESTIC_MIG_2012','DOMESTIC_MIG_2013',
 'DOMESTIC_MIG_2014','DOMESTIC_MIG_2015','DOMESTIC_MIG_2016',
 'DOMESTIC_MIG_2017','DOMESTIC_MIG_2018','DOMESTIC_MIG_2019']
dom_mig = pop.filter(dom_mig_list)

# remove DOMESTIC_MIG from columns
dom_mig.columns = [col.replace('DOMESTIC_MIG_','') for col in dom_mig.columns]

# reduce into rows
dom_mig = pd.melt(dom_mig,id_vars=['FIPStxt','State','Area_Name'],
                 var_name = 'Year',value_name = 'Dom_Mig')
```

```python
pop_df = pd.merge(dom_mig,int_mig,how = 'inner',
                  on=['FIPStxt','State','Area_Name','Year'])
```

```python
pop_df = pop_df.merge(nat_inc,how = 'inner',
                      on=['FIPStxt','State','Area_Name','Year'])
```

```python
pop_df = pop_df.merge(death,how = 'inner',
                      on=['FIPStxt','State','Area_Name','Year'])
```

```python
pop_df = pop_df.merge(birth,how = 'inner',
                      on=['FIPStxt','State','Area_Name','Year'])
```

```python
pop_df = pop_df.merge(n_pop,how = 'inner',
                      on=['FIPStxt','State','Area_Name','Year'])
```

```python
pop_df = pop_df.merge(pop_est,how = 'inner',
                      on=['FIPStxt','State','Area_Name','Year'])
```

```python
pop_df.sort_values(['FIPStxt','Year'],inplace=True)
```
