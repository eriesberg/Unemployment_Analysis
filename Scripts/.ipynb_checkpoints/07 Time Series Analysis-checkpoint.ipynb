# Dickey-Fuller test for Stationarity

```python
from statsmodels.tsa.stattools import adfuller

def dickey_fuller(timeseries):
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

dickey_fuller(data['Value'])
```

The test statistic is higher, so we need to do some differencing to create a stationary dataset

# Differencing

Using the logarithmic transformation will be the most effective with this dataset.

```python
data_diff = pd.DataFrame()
data_diff['Value'] = data['Value'].apply(np.log)
```

```python
data_diff.dropna(inplace=True)
```

```python
plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```

```python
#Re-conduct Dickey_Fuller

dickey_fuller(data_diff['Value'])
```

```python
data_diff.head(10)
```

```python
type(data_diff)
```

```python
decomposition = sm.tsa.seasonal_decompose(data_diff, model='additive')
```

```python
decomposition.plot()
plt.show()
```

### Perform additional differencing.

```python
data_diff_2 = data_diff.diff(axis=0, periods = 1)
```

```python
data_diff_2.dropna(inplace=True)
```

```python
data_diff_2
```

```python
plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff_2)
```

# Autocorrelation of Data

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(data_diff_2)
plt.show()
```
