I need to make all of the GeoFIPS codes in `df` 5 characters long.  I will use a loop to do so.

```python
df_list = df['GeoFIPS'].to_list()
for i in range(len(df_list)):
    if len(df_list[i])<5:
        df_list[i] = (5-len(df_list[i]))*'0' + df_list[i]
        
```

```python
df['GeoFIPS'] = df_list
```
