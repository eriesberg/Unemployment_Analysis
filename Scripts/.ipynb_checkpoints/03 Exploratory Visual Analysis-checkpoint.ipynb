After this check, it appears that I need to control all of the variables for county population, in order to get a better sense of what I'm dealing with.  Any correlation that may have been indicated eariler is now eliminated after controlling for population.

```python
# Divide all columns by a population estimate, 
# to get a rate that has been controlled by population

pop_control_columns = ['Unemployed', 'Employed', 'Labor_Force',
                       'Dom_Mig', 'Int_Mig', 'Nat_Inc', 'Deaths',
                       'Births','Pop_Change','gov_retire_disability',
                       'gov_medical', 'gov_income', 'gov_unemployment',
                       'gov_veterans','gov_education', 'gov_other']

for pop_control in pop_control_columns:
    df[pop_control] = df[pop_control]/df['Pop_Est']
```

```python
df
```

```python
# Create a subgroup to examine correlations

sub = df[['Unemployed','gov_medical','gov_retire_disability',
         'gov_income','gov_unemployment','gov_veterans','gov_education',
         'gov_other']]
```

```python
# Create subplot to put heatmap on
f,ax = plt.subplots(figsize=(10,10))
```
