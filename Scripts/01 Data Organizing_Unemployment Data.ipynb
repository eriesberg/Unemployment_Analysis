# Shape DataFrame and Delete Columns

```python
unemp.columns.tolist()
```

```python
# Remove Rate Columns (can be re-calculated later)
unemp = unemp[unemp.columns.drop(list(unemp.filter(regex='Unemployment_rate')))]
```

```python
# Remove _code_
unemp = unemp[unemp.columns.drop(list(unemp.filter(regex='_code_')))]
```

```python
# Remove remaining unnecessary columns
unemp = unemp[unemp.columns.drop(['Metro_2013','Median_Household_Income_2019',
 'Med_HH_Income_Percent_of_State_Total_2019'])]
```

```python
#Unemployed
unemployed_list = ['FIPS_Code','State','Area_name']
o = ['Unemployed_' + str(x) for x in range(2010,2020)]
unemployed_list += o
unemployed = unemp.filter(unemployed_list)

# remove Deaths_ from columns
unemployed.columns = [col.replace('Unemployed_','') for col in unemployed.columns]

# reduce into rows
unemployed = pd.melt(unemployed,id_vars=['FIPS_Code','State','Area_name'],
                 var_name = 'Year',value_name = 'Unemployed')
```

```python
#Employed
employed_list = ['FIPS_Code','State','Area_name']
o = ['Employed_' + str(x) for x in range(2010,2020)]
employed_list += o
employed = unemp.filter(employed_list)

# remove Employed_ from columns
employed.columns = [col.replace('Employed_','') for col in employed.columns]

# reduce into rows
employed = pd.melt(employed,id_vars=['FIPS_Code','State','Area_name'],
                 var_name = 'Year',value_name = 'Employed')
```

```python
#Labor Force
labor_list = ['FIPS_Code','State','Area_name']
o = ['Civilian_labor_force_' + str(x) for x in range(2010,2020)]
labor_list += o
labor = unemp.filter(labor_list)

# remove Employed_ from columns
labor.columns = [col.replace('Civilian_labor_force_','') for col in labor.columns]

# reduce into rows
labor = pd.melt(labor,id_vars=['FIPS_Code','State','Area_name'],
                 var_name = 'Year',value_name = 'Labor_Force')
```

```python
unemp_df = pd.merge(unemployed,employed,how='inner',
                   on=['FIPS_Code','State','Area_name','Year'])
```

```python
unemp_df = unemp_df.merge(labor,how='inner',
                   on=['FIPS_Code','State','Area_name','Year'])
```
